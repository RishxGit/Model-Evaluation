```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
Data = load_breast_cancer(as_frame=True).frame
```

```python
Data
```

```python
X,y = load_breast_cancer(return_X_y= True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train_scaled,y_train)
```

```python
print(knn.score(X_test_scaled, y_test))
```

FEATURE ENCODING

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car', as_frame=True).frame
```

```python
data
```

```python
columns_to_encode = ['lug_boot', 'safety']
encoder = OrdinalEncoder(
    categories = [ ['small','med', 'big'], #0, 1, 2. 
                     ['low','med', 'high'] 
                   ] 
) 
data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
# Print exact values to find hidden spaces or typos
print("Unique lug_boot:", data['lug_boot'].unique())
print("Unique safety:", data['safety'].unique())
```

```python
data
```

```python
encoder.inverse_transform(data[columns_to_encode])
```

```python
data = fetch_openml('adult', as_frame=True).frame
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data.occupation.value_counts()
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_values = encoder.fit_transform(data[['occupation','race']])
new_columns = encoder.get_feature_names_out(['occupation','race'])
```

```python
encoded_values
```

```python
new_columns
```

```python
# 1. Convert the numpy array into a structured pandas DataFrame
encoded_df = pd.DataFrame(encoded_values, columns=new_columns, index=data.index)

# 2. Drop the old string columns from your original data
data = data.drop(columns=['occupation', 'race'])

# 3. Concatenate the new one-hot columns back into your main dataset
data = pd.concat([data, encoded_df], axis=1)
```
